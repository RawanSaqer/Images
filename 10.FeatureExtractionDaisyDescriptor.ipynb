## Feature Extraction using Daisy Descriptor 
This local image descriptor is based on gradient orientation histograms similar to the SIFT descriptor. It is formulated in a way that allows for fast dense extraction.

https://scikit-image.org/docs/0.12.x/auto_examples/features_detection/plot_daisy.html

```python
import skimage
from skimage import color
from skimage.feature import daisy
from skimage import data
from skimage.transform import resize

import matplotlib.pyplot as plt
```

### Loading an image

#### NOTE: Daisy descriptors can only be applied on single channel images

```python
horse_gray = color.rgb2gray(skimage.io.imread('datasets/images/horse.jpg'))
```

As the loaded image is a larger one reducing the size

```python
horse = resize(horse_gray, output_shape = (300, 450), mode='reflect', anti_aliasing=False )

plt.figure(figsize=(10, 7))
plt.imshow(horse, cmap='gray');
```

### Extract DAISY feature descriptors 
* Input image = horse
* step = distance between descriptor sampling points
* radius = Radius (in pixels) of the outermost ring
* rings = Number of rings
* histograms = Number of histograms sampled per ring
* orientations = Number of orientations (bins) per histogram
* visualize = Generate a visualization of the DAISY descriptors

Returns:

* descs_array = grid of DAISY descriptors for the given image
* descs_image = Visualization of the DAISY descriptors.

### Defining the descriptors

```python
descs_array, descs_image = daisy(horse, step=150, radius=30, visualize=True)
```

### Calculate the number of the descriptors

```python
descs_num = descs_array.shape[0] * descs_array.shape[1]

descs_num
```

```python
plt.figure(figsize=(10, 7))

plt.title('DAISY descriptors', fontsize=18)
plt.imshow(descs_image);
```

### Plotting a image with descriptor with different step size and radius

Defining the descriptors

```python
descs_array_1, descs_image_1 = daisy(horse, step=100, radius=30, visualize=True)
```
