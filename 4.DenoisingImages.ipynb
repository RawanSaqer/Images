```python
import warnings
warnings.filterwarnings("ignore")
```

```python
from skimage import io
from skimage.util import random_noise
from skimage.transform import resize

import matplotlib.pyplot as plt
```

### Read the image

```python
horse = io.imread('datasets/images/horse.jpg')

plt.figure(figsize = (10, 10))
plt.imshow(horse)
```

```python
horse = resize(horse, (int(horse.shape[0]/10), int(horse.shape[1]/10)))

plt.figure(figsize = (10, 10))
plt.imshow(horse)
```

<b>random_noise</b> - Function to add random noise of various types to a floating-point image.

```python
sigma = 0.355
noisy_horse = random_noise(horse, var = sigma**2)

plt.figure (figsize = (10, 10))
plt.title("Image with added noise")
plt.imshow(noisy_horse);
```

```python
sigma = 0.155
noisy_horse = random_noise(horse, var = sigma**2)

plt.figure (figsize = (10, 10))
plt.title("Image with added noise")
plt.imshow(noisy_horse);
```

```python
from skimage.restoration import denoise_tv_chambolle, denoise_bilateral
from skimage.restoration import denoise_wavelet, estimate_sigma
```

#### Estimate the average noise standard deviation across color channels.
Robust wavelet-based estimator of the (Gaussian) noise standard deviation.

```python
sigma_est = estimate_sigma(noisy_horse,
                           multichannel=True,
                           average_sigmas=True)
sigma_est
```

<b>denoise_tv_chambolle</b> - Perform total-variation denoising on n-dimensional images.

```python
plt.figure(figsize = (10, 10))
plt.imshow(denoise_tv_chambolle(noisy_horse,
                                weight=0.1,
                                multichannel=True));
```

<b>denoise_bilateral</b> - Denoise image using bilateral filter

```python
plt.figure(figsize = (10, 10))
plt.imshow(denoise_bilateral(noisy_horse,
                             sigma_color=0.05,
                             sigma_spatial=2,
                             multichannel=True));
```

<b>denoise_wavelet</b> - Perform wavelet denoising on an image

```python
plt.figure(figsize = (10, 10))
plt.imshow(denoise_wavelet(noisy_horse,
                           multichannel=True));
```
