```python
from skimage import io
import numpy as np
import cv2

from matplotlib import pyplot as plt
```

```python
import albumentations as albu
```

### Loading an image

```python
image = io.imread('datasets/images/taj_mahal_india.jpg')
```

https://www.pexels.com/photo/taj-mahal-india-1603650/

```python
plt.figure(figsize=(12, 10))
plt.imshow(image)
plt.show()
```

### Plotting a defining function

```python
def augment_and_show(aug, image):
    image = aug(image=image)['image']
    
    plt.figure(figsize=(12, 10))
    plt.imshow(image)
```

### Rain Effect

` p (float)` – probability of applying the transform. Default: 0.5

```python
img_rain = albu.RandomRain(p=1, brightness_coefficient=0.9, drop_width=3, blur_value=1)

augment_and_show(img_rain, image)
```

### Snow effect

```python
img_snowfall = albu.RandomSnow(p=1, brightness_coeff=1.9, snow_point_lower=0.3, snow_point_upper=0.5)

augment_and_show(img_snowfall, image)
```

### Sun flare effect

```python
img_sunflare = albu.RandomSunFlare(p=1, flare_roi=(0, 0, 0.5, 0.5), angle_lower=0.6)

augment_and_show(img_sunflare, image)
```


### Fog effect

```python
img_fog = albu.RandomFog(p=1, fog_coef_lower=0.5, fog_coef_upper=0.8, alpha_coef=0.1)

augment_and_show(img_fog, image)
```
